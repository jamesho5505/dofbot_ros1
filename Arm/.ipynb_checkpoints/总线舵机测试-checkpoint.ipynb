# 舵机驱动控制

```python
import smbus
import time
#https://blog.csdn.net/Jacob_zhj/article/details/90518363
#使能I2c0的方法 https://blog.csdn.net/ManWZD/article/details/103147985?utm_medium=distribute.pc_aggpage_search_result.none-task-blog-2~all~first_rank_v2~rank_v25-1-103147985.nonecase

class Arm_Device(object):
    
    def __init__(self):
        self.bus = smbus.SMBus(1)
        self.addr = 0x15

    # 设置总线舵机角度接口：id: 1-6(0是群发) angle: 0-180  表示900 3100   0 - 180
    def Arm_serial_servo_write(self, id, angle):
        addr = 0x15
        if id == 1:
            pos = int((3100-900)*(angle-0)/(180-0) + 900)
            pos = ((pos << 8) & 0xff00) | ((pos >> 8) & 0xff) 
            self.bus.write_word_data(addr, 0x11, pos & 0xffff)
        elif id == 2:
            pos = int((3100-900)*(angle-0)/(180-0) + 900)
            pos = ((pos << 8) & 0xff00) | ((pos >> 8) & 0xff) 
            self.bus.write_word_data(addr, 0x12, pos & 0xffff)
        elif id == 3:
            angle = 180 - angle
            pos = int((3100-900)*(angle-0)/(180-0) + 900)
            pos = ((pos << 8) & 0xff00) | ((pos >> 8) & 0xff) 
            self.bus.write_word_data(addr, 0x13, pos & 0xffff)
        elif id == 4:
            angle = 180 - angle
            pos = int((3100-900)*(angle-0)/(180-0) + 900)
            pos = ((pos << 8) & 0xff00) | ((pos >> 8) & 0xff) 
            self.bus.write_word_data(addr, 0x14, pos & 0xffff)
        elif id == 5:
            pos = int((3100-900)*(angle-0)/(180-0) + 900)
            pos = ((pos << 8) & 0xff00) | ((pos >> 8) & 0xff) 
            self.bus.write_word_data(addr, 0x15, pos & 0xffff)
        elif id == 6:
            pos = int((3100-900)*(angle-0)/(180-0) + 900)
            pos = ((pos << 8) & 0xff00) | ((pos >> 8) & 0xff) 
            self.bus.write_word_data(addr, 0x16, pos & 0xffff)
        elif id == 0: # 此为所有舵机控制
            pos = int((3100-900)*(angle-0)/(180-0) + 900)
            pos = ((pos << 8) & 0xff00) | ((pos >> 8) & 0xff) 
            self.bus.write_word_data(addr, 0x17, pos & 0xffff)

    #读取指定舵机角度，返回0-180
    def Arm_serial_servo_read(self, id):
        addr = 0x15
        self.bus.write_byte(addr, id + 0x30)
        time.sleep(0.001)
        pos = self.bus.read_word_data(addr, id + 0x30)
        print(pos)
        pos = (pos >> 8 & 0xff) | (pos << 8 & 0xff00)
        print(pos)
        pos = int((180-0)*(pos-900)/(3100-900) + 0)
        print(pos)

    #读取硬件版本号
    def Arm_get_hardversion(self):
        addr = 0x15
        self.bus.write_byte(addr, 0x01)
        value = self.bus.read_word_data(addr, 0x01)
        bigver = (value & 0xff)
        littlever = (value >> 8 & 0xff)
        version = str(bigver) + '.' + str(littlever)
        print(version)

    #扭矩开关
    def Arm_serial_set_torque(self, onoff):
        addr = 0x15
        if onoff == 1:
            self.bus.write_byte_data(addr, 0x1A, 0x01)
        else:
            self.bus.write_byte_data(addr, 0x1A, 0x00)

    #设置总线舵机的编号
    def Arm_serial_set_id(self, id):
        addr = 0x15
        self.bus.write_byte_data(addr, 0x18, id&0xff)
    
    #设置编程灯颜色 1-7
    def Arm_Product_Select(self, index):
        self.bus.write_byte_data(addr, 0x04, index&0xff)
```

# TCP通信部分

```python
from socket import *
from time import ctime
import os 
import time
from Arm_Lib import Arm_Device

Arm = Arm_Device()

host = ''
port = 6000
bufsize = 1024
addr = (host, port)
times = 0;

tcpServer = socket(AF_INET,SOCK_STREAM)
tcpServer.bind(addr)
tcpServer.listen(5) #这里设置监听数为5(默认值),有点类似多线程。

#协议解析部分
def Analysis(socket, str):
    print(str)
    #    $04070113795#
    cmd = str[1:3]
    if cmd == "01": #获取硬件版本号
        data = "$0102100d#"
        socket.send(data.encode(encoding="utf-8"))
    elif cmd == "02":
        id = str[5:7]
        pos = Arm_serial_servo_read(id)
        checksum = (2 + 5 + id + pos)&0xff
        checksumstr = "%02x" % checksum
        posstr = "%02d" % pos
        data = "$0205" + str[5:7] + posstr + checksumstr + "#"
        socket.send(data.encode(encoding="utf-8"))
    elif cmd == "03":
        pos = [0,0,0,0,0,0]
        for i in range(0, 6):
            pos[i] = Arm.Arm_serial_servo_read(i+1)
        checksum = (3 + 20 + 6 + pos[0] + pos[1] + pos[2] + pos[3] + pos[4] + pos[5])&0xff
        pos1str = "%03d" % pos[0]
        pos2str = "%03d" % pos[1]
        pos3str = "%03d" % pos[2]
        pos4str = "%03d" % pos[3]
        pos5str = "%03d" % pos[4]
        pos6str = "%03d" % pos[5]
        checksumstr = "%02x" % checksum
        data = "$0320" + str[5:7] + pos1str+pos2str+pos3str+pos4str+pos5str+pos6str + checksumstr + "#"
        print (data)
        socket.send(data.encode(encoding="utf-8"))
    elif cmd == "04":  #设置指定舵机位置
        id = int(str[5:7])
        angle = int(str[7:10])
        Arm.Arm_serial_servo_write(id, angle, 500)
    elif cmd == "05":  #设置一组舵机位置
        #    $052006000180000180000180XX#
        idmax = int(str[5:7])
        if idmax == 6:
            angle1 = int(str[7:10])
            angle2 = int(str[10:13])
            angle3 = int(str[13:16])
            angle4 = int(str[16:19])
            angle5 = int(str[19:22])
            angle6 = int(str[22:25])
            Arm.Arm_serial_servo_write6(angle1, angle2, angle3, angle4, angle5, angle6)
    elif cmd == "06":  #设置舵机位置
        id = int(str[5:7])
        if id < 1 or id > 250:
            socket.send("$0602020A#".encode(encoding="utf-8")) #编号错误
            return
        Arm.Arm_serial_set_id(id)    
        socket.send("$06020008#".encode(encoding="utf-8")) #编号错误
        Arm.Arm_serial_servo_write6(90, 90, 90, 90, 90, 90)
    elif cmd == "07":  #归中舵机位置
        Arm.Arm_serial_servo_write6(90, 90, 90, 90, 90, 90)
    elif cmd == "09":  #急停|恢复舵机，关闭舵机
        state = int(str[5:6])
        Arm.Arm_serial_set_torque(state)
    elif cmd == "0A":  #设置舵机中值id:0 恢复默认  id：1-20 正常舵机中位值
        id = int(str[5:6])
        Arm.Arm_serial_servo_write_offset_switch(id)        
        
while True:
    times = 0
    print('Waiting for connection...')
    tcpClient,addr = tcpServer.accept() #拿到5个中一个监听的tcp对象和地址
    print('[+]...connected from:',addr)

    while True:
        if times == 0:
            data = "$0102100d#"#cmdResult if (not cmdStatus) else "ERROR COMMAND"
            tcpClient.send(data.encode(encoding="utf-8"))
            times = 1
        cmd = tcpClient.recv(bufsize).decode(encoding="utf-8") 
        if not cmd:
            break
        print('   [-]cmd:', cmd, len(cmd))
        index1 = cmd.find("$")
        index2 = cmd.find("#")
        if index1 >= 0 and index2 > index1:
            Analysis(tcpClient, cmd[index1:index2+1])
    tcpClient.close() #
    print(addr,'End')
tcpServer.close() #两次关闭，第一次是tcp对象，第二次是tcp服务器
```
